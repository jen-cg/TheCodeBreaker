### Translation Ciphers
- In most cases these ciphers **will** have ~26 characters
- These ciphers **will** likely have letter frequencies that match plain english very well
- These ciphers **will** have an incidence of coincidence of ~0.066 for english text 
- These ciphers **will not** yeild peaks on a periodic incidence of coincidence calculation
1. **Rail Fence**

```Python 
Break_RailFence(cipherText)
```


### Simple Monoalphabetic Substitutions
- In most cases these ciphers will have ~26 characters
    - Baconian ciphers (that have not been disguised) will only have 2
    - Text written in morse code will have two characters plus a separator character (often a space)
        - Text written in morse code may have been encrypted via an additional second method
    - At the very most a 5x5 Polybius Square Cipher can have 10 characters, but a typical encryption will only have 5
    - At the very most a 6x6 Polybius Square Cipher can have 12 characters, but a typical encryption will only have 6

- These ciphers likely **will not** have letter frequencies that match plain english
- These ciphers **will** have an incidence of coincidence of ~0.066 for english text 
- These ciphers **will not** yeild peaks on a periodic incidence of coincidence calculation

1. **Atbash**

```Python
Break_Atbash(cipherText)
```

2. **Caesar Shift**

### Polyalphabetic Substitutions
- These ciphers likely **will not** have letter frequencies that match plain english (~ flat letter frequency distribution)
- These ciphers **will** have an incidence of coincidence *less* than 0.066 for english text 
- These ciphers **will** yeild peaks on a periodic incidence of coincidence calculation

1. **Vigenere Cipher**  
Start by applying the technique using the period of the first peak of the Periodic IOC Calculation

```Python 
Break_VigenereChiSquared( cipherText, period)

Break_VigenereDictionarySearch(cipherText, period)

```

2. **Hill Cipher**  
Start by identifying the key size. 
    - The length of the text must be a multiple of the key size
    - The peaks of the Period IOC graph may coincide with a multiple of the key size
        - The peaks of the Period IOC graph may not be as extreme as with other polyalphabetic ciphers like Vigenere

Try `Break_2x2HillCrib` first 
```Python 
key, plainText = Break_2x2HillBruteForce( cipherText )
key, plainText = Break_2x2HillBigramMatching( cipherText,b1,m1,b2,m2)
key, plainText = Break_2x2HillCrib( cipherText, crib)
```
