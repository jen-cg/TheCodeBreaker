## Part 3) ngram Frequencies

```python
def ngramFrequencies(sampleText, n):
    '''
    Get the ngram frequencies of a given text 
        - ie n = 2 for bigram frequencies
        
    -Please remove all non-text characters before using
    '''

    ngramFreq = {}
    cnt = 0
    for i in range(len(sampleText) - n-1):
        cnt += 1
        ngram = ''.join([ c.upper() for c in list(sampleText[i:i + n]) ])
        
        if ngram in ngramFreq.keys():
             ngramFreq[ngram] += 1
        else:
            ngramFreq[ngram] = 1
    
    return { k: v / cnt for k, v  in sorted(ngramFreq.items(), key = lambda item: item[1], reverse=True) }
```

```python
import re
```

```python
wordList = [ 'AND', 'THAT', 'OUR', 'WHAT', 'WITH', 'THE' ]
```

```python
commonWords = open('Resources/SampleTexts/commonEnglishWords.txt').read()
commonWord = commonWords.split('\n')

wordList = []
for word in commonWord[:20][::-1]:
    if len(word) >= 3:
        wordList.append(word.upper())
        
print(wordList)
```

```python
text = open('NationalCipherChallenge2021/Challenge4B_solution.txt').read()

for word in wordList:
    result = [_.start() for _ in re.finditer(word, text)] 
    for i, ind in enumerate(result):
        if text[ind+i*2] != ' ' or text[ind+i*2+len(word)] != ' ':
            text = text[0:ind+i*2] + ' ' + text[ind+i*2:ind+i*2+len(word)] + ' ' + text[ind+i*2+len(word):]
            
text
```
